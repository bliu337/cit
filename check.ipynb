```python
import textbase, httpx, io


r = httpx.get("https://raw.githubusercontent.com/iconclass/cit/refs/heads/main/data/CATALOG.dmp")
if r.status_code == 200:
    CATALOG = textbase.TextBase(r.content)
    
```

```python
print(f"There are {len(CATALOG)} entries in the CATALOG.DMP") 

collections = {}
for entry in CATALOG:
    collection = entry.get("COL", [None])[0]
    collections.setdefault(collection, []).append(entry)   

for k, v in collections.items():
    print(f"{k}: {len(v)}")

print(sum(len(v) for _, v in collections.items()))
```

```python

LOCATIONS = {}
for entry in CATALOG:
    location = entry.get("LOCATION.INV", [None])[0]
    if location:
        LOCATIONS.setdefault(location, []).append(entry)
for k, v in LOCATIONS.items():
    print(f"{k}: {len(v)}")
```
